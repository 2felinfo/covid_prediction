# manipulation des données manquantes

```python
data.isnull().sum()#affiche une liste des colonnes et le nombre de données manquantes, 0 ça veut dire aucune donnée est manquante
```

**On remarque qu'il y a beaucoup de données manquantes d'où la nécessité de les remplir. Aussi, il existe des colonnes dont le nombre des données présentes est très inférieur par rapport au nombre des données manquantes(exemple la colonne "rapid_flu_results" (11004/11169)), d'où pour traiter la situation on doit éliminer toutes ces colonnes**

# Suppression des colonnes inutiles

```python
NoneFeatures=['batch_date','rapid_flu_results','rapid_strep_results',
'cough_severity',                    
'sob_severity',                     
'cxr_findings',                     
'cxr_impression',                   
'cxr_label',                        
'cxr_link','days_since_symptom_onset']
for item in NoneFeatures:
    data1.drop(columns=item, inplace=True)

data1.shape #affichage du nombre de lignes et colonnes aprés suppression 
```

# Traitement des données manquantes pour les variables catégoriques

```python
category = data1.select_dtypes(include=['object']).copy()
l_category = list(category)
for item in l_category:
    print(data1[item].isnull().sum())
    
```

```python
for item in l_category[2:]:
    data1[item] = data1[item].fillna(data1[item].mode()[0])#remplir les données manquantes par la methode "most frequent"
for item in l_category:
    print(data1[item].isnull().sum())
```


**On remarque qu'aucune donnée est manquante après le remplissage, d'où on a fini nos traitements pour les variables catégoriques**

# Traitement des données numériques manquantes

```python
MissedNumericalDataFeatures = ['temperature',                      
'pulse',                            
'sys',                              
'dia',                              
'rr',                               
'sats']
for item in MissedNumericalDataFeatures:
    data1[item] = data1[item].fillna(data1[item].median())
    
```

**on a remplit les cases manquantes des données numériques par le median**

```python
from sklearn.metrics import f1_score
def bestThresshold(y_true,y_pred):
    best_thresh = None
    best_score = 0
    for thresh in np.arange(0.1, 0.9, 0.01):
        score = f1_score(y_true, np.array(y_pred)>=thresh)
        if score > best_score:
            best_thresh = thresh
            best_score = score
    return best_score , best_thresh
```
